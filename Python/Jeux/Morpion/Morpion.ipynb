```python
import pygame
import sys

# Initialisation de Pygame
pygame.init()

# Paramètres de la fenêtre
WIDTH = 300
HEIGHT = 300
LINE_WIDTH = 10
BOARD_ROWS = 3
BOARD_COLS = 3
SQUARE_SIZE = WIDTH // BOARD_COLS

# Couleurs
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)

# Création de la fenêtre
window = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Tic Tac Toe")

# Initialisation du tableau de jeu
board = [['' for _ in range(BOARD_COLS)] for _ in range(BOARD_ROWS)]

# Fonction pour dessiner le tableau de jeu
def draw_board():
    window.fill(WHITE)
    for i in range(1, BOARD_ROWS):
        pygame.draw.line(window, BLACK, (0, i * SQUARE_SIZE), (WIDTH, i * SQUARE_SIZE), LINE_WIDTH)
    for j in range(1, BOARD_COLS):
        pygame.draw.line(window, BLACK, (j * SQUARE_SIZE, 0), (j * SQUARE_SIZE, HEIGHT), LINE_WIDTH)

# Fonction pour dessiner les symboles (X ou O)
def draw_symbols():
    for row in range(BOARD_ROWS):
        for col in range(BOARD_COLS):
            if board[row][col] == 'X':
                pygame.draw.line(window, BLACK, (col * SQUARE_SIZE + SQUARE_SIZE * 0.1, row * SQUARE_SIZE + SQUARE_SIZE * 0.1), 
                                 ((col + 1) * SQUARE_SIZE - SQUARE_SIZE * 0.1, (row + 1) * SQUARE_SIZE - SQUARE_SIZE * 0.1), 5)
                pygame.draw.line(window, BLACK, ((col + 1) * SQUARE_SIZE - SQUARE_SIZE * 0.1, row * SQUARE_SIZE + SQUARE_SIZE * 0.1),
                                 (col * SQUARE_SIZE + SQUARE_SIZE * 0.1, (row + 1) * SQUARE_SIZE - SQUARE_SIZE * 0.1), 5)
            elif board[row][col] == 'O':
                pygame.draw.circle(window, BLACK, (col * SQUARE_SIZE + SQUARE_SIZE // 2, row * SQUARE_SIZE + SQUARE_SIZE // 2), SQUARE_SIZE // 3, 3)

# Fonction pour vérifier s'il y a un gagnant
def check_winner(symbol):
    # Vérifier les lignes et les colonnes
    for i in range(BOARD_ROWS):
        if all(board[i][j] == symbol for j in range(BOARD_COLS)):
            return True
        if all(board[j][i] == symbol for j in range(BOARD_COLS)):
            return True
    # Vérifier les diagonales
    if all(board[i][i] == symbol for i in range(BOARD_ROWS)):
        return True
    if all(board[i][BOARD_COLS - 1 - i] == symbol for i in range(BOARD_ROWS)):
        return True
    return False

# Fonction pour vérifier si la grille est pleine
def check_draw():
    return all(board[i][j] != '' for i in range(BOARD_ROWS) for j in range(BOARD_COLS))

# Fonction pour afficher le message de fin de jeu
def draw_end_game_message(message):
    font = pygame.font.Font(None, 32)
    text = font.render(message, True, RED)
    text_rect = text.get_rect(center=(WIDTH // 2, HEIGHT // 2))
    window.blit(text, text_rect)
    pygame.display.flip()

# Variable pour garder la trace de l'actuel joueur
current_player = 'X'

# Boucle de jeu
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()
            sys.exit()
        if event.type == pygame.MOUSEBUTTONDOWN:
            mouseX = event.pos[0]
            mouseY = event.pos[1]
            clicked_row = mouseY // SQUARE_SIZE
            clicked_col = mouseX // SQUARE_SIZE
            if board[clicked_row][clicked_col] == '':
                board[clicked_row][clicked_col] = current_player
                if check_winner(current_player):
                    draw_board()
                    draw_symbols()
                    draw_end_game_message(f'Player {current_player} wins!')
                    running = False
                elif check_draw():
                    draw_board()
                    draw_symbols()
                    draw_end_game_message('Draw!')
                    running = False
                else:
                    current_player = 'O' if current_player == 'X' else 'X'

    draw_board()
    draw_symbols()
    pygame.display.update()
```
